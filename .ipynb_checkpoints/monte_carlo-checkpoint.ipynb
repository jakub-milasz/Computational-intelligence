# Implementacja algorytmu Monte Carlo

Metoda Monte Carlo służy do modelowania matematycznego złożonych procesów, np. obliczania całek. Polega na wylosowaniu N punktów (x, y) i porównaniu położenia każdego punktu, czy jest poniżej, czy powyżej danej funkcji. Na końcu naszym wynikiem jest iloraz liczby punktów pod wykresem i N, pomnożony przez długość i wysokość badanego obszaru (granice całkowania).

```python
def monte_carlo_visualize(number_of_points, x_coords, y_coords, x_success, y_success):
    success = 0
    for i in range(number_of_points):
        x = np.random.uniform(0, 1.1)
        y = np.random.uniform(0, 1.1)
        x_coords.append(x)
        y_coords.append(y)
        if x**2 + y**2 <= 1:
            success += 1
            x_success.append(x)
            y_success.append(y)
    return 4*success/number_of_points*1.1*1.1
```

```python
def monte_carlo(number_of_points):
    success = 0
    for i in range(number_of_points):
        x = np.random.uniform(0, 1.1)
        y = np.random.uniform(0, 1.1)
        if x**2 + y**2 <= 1:
            success += 1
    return 4*success/number_of_points*1.1*1.1
```

```python
def monte_carlo_moment(number_of_points):
    success = 0
    res = []
    for i in range(1, number_of_points+1):
        x = np.random.uniform(0, 1.1)
        y = np.random.uniform(0, 1.1)
        if x**2 + y**2 <= 1:
            success += 1
        res.append(4*success/i*1.1*1.1)
    return res
```

# Wizualizacja ćwiartki okręgu z wylosowanymi punktami

```python
N = [100, 1000, 10000, 100000]
x_coords = []
y_coords = []
x_success = []
y_success = []
x_circle = np.linspace(0, 1, 200)
fig, axes = plt.subplots(2, 2, figsize = (10, 10))
for i, ax in enumerate(axes.flat):
    pi = monte_carlo_visualize(N[i], x_coords, y_coords, x_success, y_success)
    ax.scatter(x_coords, y_coords, c='blue', s=0.3)
    ax.scatter(x_success, y_success, c='red', s=0.3)
    ax.plot(x_circle, [np.sqrt(1 - x**2) for x in x_circle], c='green')
    ax.legend(title=f"PI = {pi:.4f}", loc="lower left")
    ax.set_title(f'Wylosowane punkty N = {N[i]}')
```

# Wizualizacja za pomocą wykresów pudełkowych

Dla każdej ilości punktów obliczymy po 10 razy wartość liczby PI i zobaczymy, jak będzie wyglądać zróżnicowanie wyników w zależności od liczby wylosowanych punktów.

```python
results = {n: [] for n in N}

for n in N:
    for _ in range(10):
        results[n].append(monte_carlo(n))

data = pd.DataFrame({key: value for key, value in results.items()})
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, color='white')
plt.axhline(y=3.14159265359, color='r', linestyle='--', linewidth=1.5)
plt.xlabel("Liczba punktów")
plt.ylabel("Przybliżona wartość PI")
plt.title("Rozkład estymacji liczby pi dla różnej liczby punktów")
plt.show()
```

Na powyższym wykresie możemy zauważyć, że wraz ze wzrostem liczby losowanych punktów, maleje rozbieżność uzyskiwanych wyników. Zatem dzięki większej ilości punktów mamy większe szanse na dokładniejszy wynik oraz nasz błąd będzie minimalny. Wiąże się to jednak z dłuższym czasem wykonywania algorytmu. Minimalizacja błędu wraz ze wzrostem ilości punktów wynika również ze wzoru na błąd metody Monte Carlo, który ma postać $\frac{1}{\sqrt{N}}$.

# Chwilowa estymowana wartość PI

Przyjrzyjmy się, jak wyglądają ścieżki dążenia do wartości liczby PI. W tym celu wylosujemy 10 razy po 10000 punktów i zobaczymy, jak estymowana wartość liczby PI zmienia się na bieżąco.

```python
plt.figure(figsize=(10, 5))
for i in range(10):
    pis = monte_carlo_moment(10000)
    x = np.arange(1, 10001, 1)
    plt.plot(x, pis, linewidth=0.5, c='blue')
    plt.axhline(y=3.14159265359, color='r', linestyle='--', linewidth=0.8)
    plt.title(f'Estymacja liczby PI (10 razy po 10000 punktów)')
```

Na wykresie możemy zaobserwować, że na początku wartość estymowana jest daleka od rzeczywistej wartości, jednak wraz ze wzrostem liczby wylosowanych punktów zbliża się do niej. Ponadto, możemy zauważyć, że przeprowadzone 10 symulacji w początkowej fazie wykresu mają zróżnicowane ścieżki, a w późniejszej fazie ścieżki te niemal się pokrywają. Ten wykres potwierdza naszą wcześniejszą obserwację na wykresie pudełkowym związaną ze zróżnicowaniem wyników.
